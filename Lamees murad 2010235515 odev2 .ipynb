Glomerüller, böbreklerde bulunan ve kandaki atıkları filtreleyen küçük filtrelerdir. Böbrekler hasar gördüğü durumda kanı olması gerektiği gibi hızlı bir şekilde filtreleyemez ve hastalıklara neden olur. Glomerüler filtrasyon hızı (GFR) testi, böbreklerin ne kadar kan filtrelediğini ölçerek böbrek hastalıklarının kontrolünü sağlar, Glomerüller kanı süzerek kanda bulunan zararlı maddelerin vücuttan atılmasını sağlamaktadır. GFR (Glomerüler Filtrasyon Hızı) fonksiyon gösteren her bir nefronda, glomerüllerden birim zamanda süzülen plazma miktarının toplamıdır.

```python
import tkinter as tk

def hesapla_GFR_tk():
    try:
        serum_kreatinin = float(entry_kreatinin.get())
        yaş = float(entry_yaş.get())
        cinsiyet = entry_cinsiyet.get()

        if cinsiyet.upper() == "E":
            gfr = ((140 - yaş) * serum_kreatinin) / (72)
        elif cinsiyet.upper() == "K":
            gfr = ((140 - yaş) * serum_kreatinin) / (72 * 0.85)
        else:
            label_sonuc["text"] = "Geçersiz cinsiyet girişi! Lütfen 'E' veya 'K' girin."
            return

        label_sonuc["text"] = f"GFR Değeri: {gfr} mL/dk/1.73m²"
    except ValueError:
        label_sonuc["text"] = "Geçersiz değer! Lütfen sayısal bir değer girin."

def hakkında():
    about_window = tk.Toplevel(root)
    about_window.title("Hakkında")
    about_window.geometry("300x200")
    about_window.resizable(False, False)

    label_info = tk.Label(about_window, text="Bu program GFR hesaplama için yapılmıştır.\nÖğrenci Bilgileri:\nÖğrenci Adı: [Adınız]\nÖğrenci Numarası: [Numaranız]")
    label_info.pack()

root = tk.Tk()
root.title("GFR Hesaplama")

label_kreatinin = tk.Label(root, text="Serum Kreatinin Değeri (mg/dL):")
label_kreatinin.grid(row=0, column=0)
entry_kreatinin = tk.Entry(root)
entry_kreatinin.grid(row=0, column=1)

label_yaş = tk.Label(root, text="Yaşınız:")
label_yaş.grid(row=1, column=0)
entry_yaş = tk.Entry(root)
entry_yaş.grid(row=1, column=1)

label_cinsiyet = tk.Label(root, text="Cinsiyet (E/K):")
label_cinsiyet.grid(row=2, column=0)
entry_cinsiyet = tk.Entry(root)
entry_cinsiyet.grid(row=2, column=1)

button_hesapla = tk.Button(root, text="GFR Hesapla", command=hesapla_GFR_tk)
button_hesapla.grid(row=3, column=0, columnspan=2)

label_sonuc = tk.Label(root, text="")
label_sonuc.grid(row=4, column=0, columnspan=2)

menu = tk.Menu(root)
root.config(menu=menu)
help_menu = tk.Menu(menu)
menu.add_cascade(label="Hakkında", menu=help_menu)
help_menu.add_command(label="Hakkında", command=hakkında)

root.mainloop()
```
